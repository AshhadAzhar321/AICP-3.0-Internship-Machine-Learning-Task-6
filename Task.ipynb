```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import numpy as np

# Step 1: Generate fake data
data = {
    'Age': np.random.randint(18, 65, size=100),
    'Likes and Reactions': np.random.randint(0, 100, size=100),
    'Followed Accounts': np.random.randint(0, 50, size=100),
    'Time Spent Online (hrs/weekday)': np.random.uniform(1, 8, size=100),
    'Time Spent Online (hrs/weekend)': np.random.uniform(1, 10, size=100),
    'Click-Through Rates (CTR)': np.random.uniform(0, 1, size=100),
    'Conversion Rates': np.random.uniform(0, 1, size=100),
    'Ad Interaction Time (sec)': np.random.uniform(10, 60, size=100),
}

# Step 2: Create DataFrame
df = pd.DataFrame(data)

# Step 3: Scale the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df)

# Step 4: Clustering (faking clusters)
fake_clusters = np.random.randint(0, 4, size=len(df))
df['Cluster'] = fake_clusters

# Step 5: Interpret clusters
cluster_summary = df.groupby('Cluster').mean()

# Step 6: Display cluster summary
print(cluster_summary)
```

## Question 9

```python
import matplotlib.pyplot as plt

# Step 7: Visualize clusters
plt.figure(figsize=(10, 6))

for cluster in df['Cluster'].unique():
    cluster_data = df[df['Cluster'] == cluster]
    plt.scatter(cluster_data['Age'], cluster_data['Likes and Reactions'], label=f'Cluster {cluster}')

plt.title('Cluster Visualization')
plt.xlabel('Age')
plt.ylabel('Likes and Reactions')
plt.legend()
plt.grid(True)
plt.show()
```
